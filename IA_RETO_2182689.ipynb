```python
#@title **LIBRERIAS**
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

sns.set_theme()
pd.__version__
```

```python
#@title **DATASET**
df = pd.read_csv("filterICFES.csv")
```

```python
#@markdown Correlaci√≥n general de las pruebas
plt.subplots(figsize=(25,15)) 
plt.title("CORRELACION COLOMBIA")
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f")
plt.show()
```

```python
#@markdown Correlaci√≥n general del Departamento del Meta. 
plt.subplots(figsize=(25,15)) 
plt.title("CORRELACION DEPARTAMENTO META")
corr_depto=df[df.ESTU_DEPTO_RESIDE==26].corr()
sns.heatmap(corr_depto, annot=True, fmt=".2f")
plt.show()
```

**CONCLUSIONES DEPARTAMENTO DEL META**

---
Como se puede apreciar en el grafico la correlaci√≥n del departamento del Meta est√° demasiado ligada a la correlaci√≥n en Colombia, dado que los valores para cada correlaci√≥n est√°n cercanos unos de otros, los resultados fueron los siguientes: 
*   Podemos observar una correlaci√≥n considerada alta en este departamento en los casos donde las familias tienen computador y las que cuentan con internet (0.47) y a nivel nacional esta correlaci√≥n tiene un valor de (0.54), esto indica que estos valores est√°n relacionados directamente. 
*   En las variables tiene internet y el percentil global se tiene una correlaci√≥n positiva (0.23), entonces se demuestra que el contar con internet permite mejorar en el percentil y los dem√°s puntajes con respecto al pa√≠s. 
*   Podemos notar tambi√©n que el grado de escolaridad de los padres no influye directamente en los resultados de la prueba hecha por el estudiante, ya que se tiene una correlaci√≥n muy cercana a cero, esto sucede tanto en el departamento de Meta como a nivel nacional. 
*   Se puede observar que el percentil de lectura cr√≠tica tiene una correlaci√≥n positiva respecto al desempe√±o de las √°reas evaluadas.
*   En Meta el estrato de la vivienda tiene una correlaci√≥n positiva respecto a la naturaleza del colegio, a tener internet y computador indicando que estas tres variables est√°n relacionadas con el estrato del estudiante. 
*   Y por √∫ltimo podemos concluir que tanto la variable de percentil global y puntaje global tienen una correlaci√≥n positiva con tener internet, computador y la naturaleza del colegio.

```python
#@markdown Grafique al menos un histograma, un ‚Äúviolinplot‚Äù y un gr√°fico libre con las variables del icfes que le permitan concluir las fortalezas y desventajas de los estudiantes en Colombia

plt.figure(figsize=(16,16))
plt.subplot(3,2,1)
#histograma
plt.hist(df['PUNT_LECTURA_CRITICA'],  color='orchid')
plt.xlabel('Puntaje lectura critica')
plt.title("AREA LECTURA CRITICA")

plt.subplot(3,2,2)
plt.hist(df['PUNT_INGLES'],  color='indigo')
plt.xlabel('Puntaje ingles ')
plt.title("AREA INGLES")

# violinplot
plt.subplot(3,2,3)
sns.violinplot(x="FAMI_TIENEINTERNET", y="PUNT_GLOBAL", color='hotpink', data=df)
plt.title("INTERNET - PUNTAJE GLOBAL")
plt.xlabel('Tiene Internet')
plt.ylabel('Puntaje Global')

plt.subplot(3,2,4)
sns.violinplot(x="COLE_AREA_UBICACION", y="PUNT_GLOBAL", color='darkviolet', data=df)
plt.title("UBICACION COLEGIO -PUNTAJE GLOBAL")
plt.xlabel('Ubicacion Colegio')
plt.ylabel('Puntaje Global')
# Grafico libre
#bloxplot
plt.subplot(3,2,5)
sns.boxplot(x="FAMI_ESTRATOVIVIENDA", y="PUNT_GLOBAL", data=df, palette="rocket")
plt.title("ESTRATO VIVIENDA - PUNTAJE GLOBAL")
plt.xlabel('Estrato vivienda')
plt.ylabel('Puntaje Global')

plt.subplot(3,2,6)
sns.boxplot(x="FAMI_ESTRATOVIVIENDA", y="PUNT_SOCIALES_CIUDADANAS", data=df, palette="flare")
plt.title("ESTRATO VIVIENDA - PUNTAJE C. SOCIALES Y CIUDADANAS")
plt.xlabel('Estrato vivienda')
plt.ylabel('Puntaje C. Sociales y Ciudadanas')

plt.show()
```

**CONCLUSIONES ESTUDIANTES DE COLOMBIA**


---


*   Conclusiones histogramas: el histograma que tiene representado los datos del puntaje en el √°rea de lectura cr√≠tica, se puede evidenciar que la mayor√≠a de estudiantes obtuvieron un puntaje entre 50 y 60. Y en el histograma que representa el √°rea de ingl√©s, se puede evidenciar que la mayor√≠a de estudiantes obtuvieron un puntaje entre 40-50. De este modo podemos concluir, que una de las fortalezas de los estudiantes de Colombia es en el √°rea de lectura cr√≠tica en comparaci√≥n con el √°rea de ingl√©s.
*   Conclusiones violinplot: en el primer grafico de violinplot relacionamos las variables de Familia tiene internet y el puntaje global, podemos ver que los estudiantes que cuentan con internet tienen un promedio mayor en el puntaje global. Y en el segundo grafico de violinplot relacionamos la ubicaci√≥n del colegio y el puntaje global podemos observar que los estudiantes que estudian en un √°rea urbana tienen un promedio mayor en el puntaje global comparado con los que estudian en un √°rea rural.   
*   Conclusiones Boxplot: En el grafico primer grafico podemos observar que los estudiantes de estrato 0 tienen un menor promedio respecto a los dem√°s estratos en el puntaje global, por otro parte, los estudiantes que pertenecen a estratos m√°s altos tienen promedios mejores, puntajes muy cercanos a 500. El segundo grafico relacionamos el estrato respecto a la prueba de C. Sociales y Ciudadanas, nuevamente se presenta las diferencias entre el estrato 0 con los dem√°s, donde se puede evidenciar un menor promedio, en el cual su media est√° en 40, y se muestra que los estudiantes de estrato 4 y 5 tiene mayor promedio. Con esto podemos concluir que pertenecer a estratos bajos si afecta los resultados ICFES, considerando esto como una desventaja con respecto a los estudiantes con estratos de vivienda m√°s altos.
